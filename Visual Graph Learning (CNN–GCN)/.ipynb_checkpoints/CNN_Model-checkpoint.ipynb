```python
# @title 20-day CNN Architecture
# (BasicBlock, Bottleneck, CNN20d)

class BasicBlock(nn.Module):
    """
    ResNet åŸºç¤æ®˜å·®å€å¡Š (å°æ‡‰åœ–ä¸­çš„ 5x3 -> 5x3)
    """
    def __init__(self, in_channels, out_channels):
        super(BasicBlock, self).__init__()

        # --- ä¸»è·¯å¾‘ (Main Path) ---
        self.main_path = nn.Sequential(
            # 1. 5x3 Conv
            nn.BatchNorm2d(in_channels),
            nn.GELU(),
            nn.Conv2d(in_channels, out_channels, kernel_size=(5, 3), stride=1, padding=(2, 1)),
            # 2. 5x3 Conv
            nn.BatchNorm2d(out_channels),
            nn.GELU(),
            nn.Conv2d(out_channels, out_channels, kernel_size=(5, 3), stride=1, padding=(2, 1)),

        )

        # --- è·³æ¥ (Shortcut Path) ---
        self.shortcut = nn.Sequential() # é»˜èªç‚º Identity

        # å¦‚æœè¼¸å…¥/è¼¸å‡ºé€šé“æ•¸ä¸åŒ (ä¾‹å¦‚ 64 -> 128)ï¼Œ
        # æˆ‘å€‘éœ€è¦ä¸€å€‹ 1x1 å·ç©ä¾†åŒ¹é…ç¶­åº¦
        if in_channels != out_channels:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_channels, out_channels, kernel_size=1, stride=1, padding=0),
                nn.BatchNorm2d(out_channels)
            )

    def forward(self, x):
        main_out = self.main_path(x)
        shortcut_out = self.shortcut(x)
        out = main_out + shortcut_out # ä¸åŠ  final activation
        return out

class BottleneckBlock(nn.Module):
    """
    ResNet ç“¶é ¸æ®˜å·®å€å¡Š (å°æ‡‰åœ–ä¸­çš„ 1x1 -> 5x3 -> 1x1)
    """
    def __init__(self, in_channels, out_channels, squeeze_channels):
        """
        Args:
            in_channels (int): è¼¸å…¥é€šé“æ•¸
            out_channels (int): è¼¸å‡ºé€šé“æ•¸ (ä¾‹å¦‚ 512)
            squeeze_channels (int): ç“¶é ¸å£“ç¸®å¾Œçš„é€šé“æ•¸ (ä¾‹å¦‚ 128)
        """
        super(BottleneckBlock, self).__init__()

        # --- ä¸»è·¯å¾‘ (Main Path) ---
        self.main_path = nn.Sequential(
            # 1. 1x1 Conv (å£“ç¸®)
            nn.BatchNorm2d(in_channels),
            nn.GELU(),
            nn.Conv2d(in_channels, squeeze_channels, kernel_size=1, stride=1, padding=0),


            # 2. 5x3 Conv (æ‚¨çš„æ ¸å¿ƒå·ç©)
            nn.BatchNorm2d(squeeze_channels),
            nn.GELU(),
            nn.Conv2d(squeeze_channels, squeeze_channels, kernel_size=(5, 3), stride=1, padding=(2, 1)),


            # 3. 1x1 Conv (æ“´å±•)
            nn.BatchNorm2d(squeeze_channels),
            nn.GELU(),
            nn.Conv2d(squeeze_channels, out_channels, kernel_size=1, stride=1, padding=0),
        )

        # --- è·³æ¥ (Shortcut Path) ---
        self.shortcut = nn.Sequential() # é»˜èªç‚º Identity

        # å¦‚æœè¼¸å…¥/è¼¸å‡ºé€šé“æ•¸ä¸åŒ (ä¾‹å¦‚ 256 -> 512)ï¼Œ
        # æˆ–é€™æ˜¯ç¬¬ä¸€å€‹å€å¡Šï¼Œæˆ‘å€‘éœ€è¦ä¸€å€‹ 1x1 å·ç©ä¾†åŒ¹é…ç¶­åº¦
        if in_channels != out_channels:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_channels, out_channels, kernel_size=1, stride=1, padding=0),
                nn.BatchNorm2d(out_channels)
            )

    def forward(self, x):
        main_out = self.main_path(x)
        shortcut_out = self.shortcut(x)
        out = main_out + shortcut_out # ä¸åŠ  final activation
        return out

class CNN20d(nn.Module):

    def init_weights(self, m):
        if isinstance(m, nn.Linear) or isinstance(m, nn.Conv2d):
            torch.nn.init.xavier_uniform_(m.weight)
            if m.bias is not None:
                m.bias.data.fill_(0.01)

    def _make_layer(self, block_type, in_channels, out_channels, num_blocks, squeeze_channels=None):
        """
        å»ºç«‹ä¸€å€‹åŒ…å« N å€‹æ®˜å·®å€å¡Šçš„å±¤ (Stage)
        """
        layers = []

        # ç¬¬ä¸€å€‹å€å¡Šï¼šè™•ç†è¼¸å…¥/è¼¸å‡ºé€šé“æ•¸ä¸åŒçš„æƒ…æ³
        if block_type == BottleneckBlock:
            layers.append(block_type(in_channels, out_channels, squeeze_channels))
        else: # BasicBlock
            layers.append(block_type(in_channels, out_channels))

        # å‰©é¤˜çš„ (num_blocks - 1) å€‹å€å¡Š
        for _ in range(1, num_blocks):
            if block_type == BottleneckBlock:
                # å¾ŒçºŒå€å¡Šçš„ in_channels ç­‰æ–¼ out_channels
                layers.append(block_type(out_channels, out_channels, squeeze_channels))
            else: # BasicBlock
                layers.append(block_type(out_channels, out_channels))

        return nn.Sequential(*layers)

    def __init__(self, num_classes=2, regression_outputs=1):
        super(CNN20d, self).__init__()

        # 1. Stem (è–å¹¹å±¤) - æ ¹æ“šæ‚¨çš„ H/W èª¿æ•´
        self.conv1 = nn.Sequential(
            # ä½¿ç”¨ 5x3 å·ç©æ ¸ (kernel_size=(5, 3)) æ›´é©åˆæ‚¨çš„éæ–¹å½¢è¼¸å…¥
            # padding=(2, 1) å¯ä»¥åœ¨ stride=1 æ™‚ä¿æŒ H å’Œ W å°ºå¯¸ä¸è®Š
            nn.BatchNorm2d(1),
            nn.GELU(),
            nn.Conv2d(1, 64, kernel_size=(5, 3), stride=1, padding=(2, 1))
        )

        self.pool1 = nn.MaxPool2d((2, 1)) # ç¬¬ä¸€å€‹ MaxPool

        # 2. Stage 1 - å°æ‡‰åœ–ä¸­ x2 (BasicBlock, 64)
        self.stage1 = self._make_layer(BasicBlock, 64, 64, num_blocks=2)

        # 3. Stage 2 - å°æ‡‰åœ–ä¸­ x2 (BasicBlock, 128)
        self.stage2 = self._make_layer(BasicBlock, 64, 128, num_blocks=2)

        # 4. Stage 3 - å°æ‡‰åœ–ä¸­ x2 (BasicBlock, 256)
        self.stage3 = self._make_layer(BottleneckBlock, 128, 256, num_blocks=2, squeeze_channels=64)
        self.pool2 = nn.MaxPool2d((2, 1)) # ç¬¬äºŒå€‹ MaxPool

        # 5. Stage 4 - å°æ‡‰åœ–ä¸­ x2 (Bottleneck, 128 -> 512)
        # ç“¶é ¸å£“ç¸®é€šé“æ•¸ç‚º 128ï¼Œè¼¸å‡ºé€šé“æ•¸ç‚º 512
        self.stage4 = self._make_layer(BottleneckBlock, 256, 512, num_blocks=2, squeeze_channels=128)


        # 6. Final Pooling (å…¨å±€å¹³å‡æ± åŒ–)
        # é€™æœƒå°‡ (Batch, 512, H, W) -> (Batch, 512, 1, 1)
        self.avgpool = nn.AdaptiveAvgPool2d((1, 1))

        # 7. Dropout å’Œå…¨é€£æ¥å±¤
        self.DropOut = nn.Dropout(p=0.5)

        # flattened_size ç¾åœ¨å›ºå®šç‚ºæœ€å¾Œçš„é€šé“æ•¸ 512
        self.classification_fc = nn.Linear(512, num_classes)
        self.classification_activation = nn.Softmax(dim=1)

        # self.regression_fc = nn.Linear(512, regression_outputs)
        # self.regression_activation = nn.Identity()

        self.apply(self.init_weights)

    def forward(self, x, task_type='classification', embedding=None, return_params=False):
        # ç¢ºä¿è¼¸å…¥æ˜¯ (Batch, Channels, H, W)
        # åŸå§‹ç¢¼æ˜¯ (Batch, H, W)ï¼Œæ‰€ä»¥ unsqueeze(1) è®Šæˆ (Batch, 1, H, W)
        x = x.unsqueeze(1).to(torch.float32)

        outputs = {}

        # æŒ‰ç…§æ¶æ§‹åœ–é †åºåŸ·è¡Œ
        x = self.conv1(x)
        x = self.pool1(x)
        outputs['conv1'] = x

        x = self.stage1(x)
        outputs['stage1'] = x

        x = self.stage2(x)
        outputs['stage2'] = x

        x = self.stage3(x)
        x = self.pool2(x)
        outputs['stage3'] = x

        x = self.stage4(x)
        outputs['stage4'] = x

        x = self.avgpool(x)
        # x shape: (Batch, 512, 1, 1)

        # å±•å¹³ (Flatten)
        x = torch.flatten(x, 1) # (Batch, 512)

        x = self.DropOut(x)
        outputs['flatten'] = x

        if embedding in outputs:
            return outputs[embedding]

        # Task-specific output (é€™éƒ¨åˆ†é‚è¼¯èˆ‡æ‚¨åŸç¢¼ç›¸åŒ)
        if task_type == 'classification':
            x = self.classification_fc(x)
            x = self.classification_activation(x)
        # elif task_type == 'regression':
        #     x = self.regression_fc(x)
        #     x = self.regression_activation(x)

        # (é€™éƒ¨åˆ†é‚è¼¯èˆ‡æ‚¨åŸç¢¼ç›¸åŒ)
        if return_params and embedding:
            return {
                'params': dict(self.named_parameters()),
                'output': outputs[embedding]
            }

        return x
```

```python
# @title Setting of Model Training

def make_tensor_dataset(X, y, mode="top_k", top_k=9, threshold=0.003):
    """
    å°‡å·²ç¶“å°é½Šçš„ X (numpy) å’Œ y (DataFrame)
    è½‰æ›æˆ CNN å¯ä»¥ç”¨çš„ TensorDataset

    æ ¹æ“š 'mode' åƒæ•¸æ±ºå®š label çš„ç”Ÿæˆæ–¹å¼ï¼š
    - mode="top_k":    (é è¨­) ä¾æ“šåŒä¸€å¤©çš„å ±é…¬ç‡æ’åï¼Œæ¨™å‡º top_k ç‚º 1ï¼Œå…¶é¤˜ç‚º 0ã€‚
    - mode="threshold": ä¾æ“š 'Return_5D' >= threshold (é è¨­ 0.003) ç‚º 1ï¼Œå…¶é¤˜ç‚º 0ã€‚
    """
    # å»ºç«‹å‰¯æœ¬é¿å…è­¦å‘Š
    y = y.copy()

    # å–å‡ºäº¤æ˜“æ—¥èˆ‡å ±é…¬ç‡
    y.columns = ['TradeDate', 'Ticker', 'Return_5D']

    if mode == "top_k":   # top_k=9
        # åœ¨æ¯å€‹æ—¥æœŸå…§ä¾å ±é…¬ç‡æ’å
        y['rank'] = y.groupby('TradeDate')['Return_5D'].rank(ascending=False, method='first')

        # æ¨™è¨˜ top_k ç‚º 1ï¼Œå…¶é¤˜ç‚º 0
        y['label'] = (y['rank'] <= top_k).astype(int)

    elif mode == "threshold":   # threshold=0.003
        # æ¨™è¨˜ 'Return_5D' >= threshold ç‚º 1ï¼Œå…¶é¤˜ç‚º 0
        y['label'] = (y['Return_5D'] >= threshold).astype(int)

    else:
        # è™•ç†æœªçŸ¥çš„ mode
        raise ValueError(f"æœªçŸ¥çš„ mode: '{mode}'ã€‚è«‹ä½¿ç”¨ 'top_k' æˆ– 'threshold'ã€‚")

    # è½‰æ›æˆ tensor
    X_tensor = torch.tensor(X, dtype=torch.float32)
    y_tensor = torch.tensor(y['label'].values, dtype=torch.long)

    dataset = TensorDataset(X_tensor, y_tensor)
    return dataset

def loss_function(output, ground_truth, alpha):
    if output.dim() == 1:
        output = output.unsqueeze(1)
    if ground_truth.dim() == 1:
        ground_truth = ground_truth.unsqueeze(1)

    reg_loss = F.mse_loss(output, ground_truth)

    # broadcasting
    output_diff = output - output.t()
    gt_diff = ground_truth - ground_truth.t()

    rank_loss = F.relu(-output_diff * gt_diff).sum() / 2

    return reg_loss + alpha * rank_loss

def plot_losses(train_losses, val_losses, mode):
    plt.figure(figsize=(10, 6))
    plt.plot(train_losses, label='Training Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.title(f'{mode.capitalize()} Training History')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

def train_model(model, train_dataset, test_dataset, mode='classification',
                num_epochs=50, patience=2, learning_rate=0.0001, batch_size=64, weight_decay=1e-4, lr_factor=0.5, lr_patience=2):
    device = torch.device('cuda')
    model = model.to(device)

    # Create data loaders
    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=False)
    test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    # Define loss function based on mode
    if mode == 'classification':
        criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
    else:  # regression
        criterion = loss_function

    # Initialize optimizer
    if mode == 'classification':
        optimizer = optim.AdamW(model.parameters(), lr=learning_rate, weight_decay=weight_decay)
    else:
        # For regression, only optimize the regression head
        optimizer = optim.AdamW(
            [{'params': model.regression_fc.parameters()}],
            lr=learning_rate,
            weight_decay=weight_decay)

    # åŠ å…¥ learning rate scheduler
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
    optimizer,
    T_max=num_epochs,  # å®Œæ•´è¨“ç·´é€±æœŸ
    eta_min=1e-6       # æœ€ä½å­¸ç¿’ç‡
    )

    # Initialize tracking variables
    history = {
        'train_loss': [], 'train_acc': [], 'train_precision': [], 'train_recall': [], 'train_f1': [],
        'val_loss': [], 'val_acc': [], 'val_precision': [], 'val_recall': [], 'val_f1': []
    }
    best_val_loss = float('inf')
    best_model = None
    best_epoch = 0
    patience_counter = 0

    # Training loop
    for epoch in tqdm(range(num_epochs)):
        model.train()
        epoch_train_loss = 0
        train_labels, train_preds = [], []

        for batch_data in train_loader:
            batch_x = batch_data[0].to(device)
            if mode == 'classification':
                batch_y = batch_data[1].to(device)  # Classification labels
            else:
                batch_y = batch_data[2].to(device)  # Regression labels

            optimizer.zero_grad()
            outputs = model(batch_x, task_type=mode)

            if mode == 'classification':
                loss = criterion(outputs, batch_y)
            else:
                loss = criterion(outputs, batch_y.unsqueeze(1),0.3)

            loss.backward()
            optimizer.step()

            epoch_train_loss += loss.item()
            # æ”¶é›†è¨“ç·´é æ¸¬çµæœ
            preds = torch.argmax(outputs, dim=1)
            train_labels.extend(batch_y.cpu().numpy())
            train_preds.extend(preds.cpu().numpy())

        # è¨ˆç®—è¨“ç·´é›†æŒ‡æ¨™
        avg_train_loss = epoch_train_loss / len(train_loader)
        epoch_train_acc = accuracy_score(train_labels, train_preds)
        epoch_train_precision = precision_score(train_labels, train_preds, average='binary', zero_division=0)
        epoch_train_recall = recall_score(train_labels, train_preds, average='binary', zero_division=0)
        epoch_train_f1 = f1_score(train_labels, train_preds, average='binary', zero_division=0)

        history['train_loss'].append(avg_train_loss)
        history['train_acc'].append(epoch_train_acc)
        history['train_precision'].append(epoch_train_precision)
        history['train_recall'].append(epoch_train_recall)
        history['train_f1'].append(epoch_train_f1)

        # Validation phase
        model.eval()
        epoch_val_loss = 0
        val_labels, val_preds = [], []

        with torch.no_grad():
            for batch_data in test_loader:
                batch_x = batch_data[0].to(device)
                if mode == 'classification':
                    batch_y = batch_data[1].to(device)  # Classification labels
                else:
                    batch_y = batch_data[2].to(device)  # Regression labels

                outputs = model(batch_x, task_type=mode)

                if mode == 'classification':
                    loss = criterion(outputs, batch_y)
                else:
                    loss = criterion(outputs, batch_y.unsqueeze(1),0.3)

                epoch_val_loss += loss.item()

                # æ”¶é›†é©—è­‰é æ¸¬çµæœ
                preds = torch.argmax(outputs, dim=1)
                val_labels.extend(batch_y.cpu().numpy())
                val_preds.extend(preds.cpu().numpy())

        # è¨ˆç®—é©—è­‰é›†æŒ‡æ¨™
        avg_val_loss = epoch_val_loss / len(test_loader)
        epoch_val_acc = accuracy_score(val_labels, val_preds)
        epoch_val_precision = precision_score(val_labels, val_preds, average='binary', zero_division=0)
        epoch_val_recall = recall_score(val_labels, val_preds, average='binary', zero_division=0)
        epoch_val_f1 = f1_score(val_labels, val_preds, average='binary', zero_division=0)

        history['val_loss'].append(avg_val_loss)
        history['val_acc'].append(epoch_val_acc)
        history['val_precision'].append(epoch_val_precision)
        history['val_recall'].append(epoch_val_recall)
        history['val_f1'].append(epoch_val_f1)

        # æ›´æ–° scheduler (ä»¥ validation loss ç‚ºä¾æ“š)
        scheduler.step()

        # Early stopping check
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            best_model = deepcopy(model.state_dict())
            best_epoch = epoch
            patience_counter = 0
        # else:
        #     patience_counter += 1

        # if patience_counter >= patience:
        #   print(f"\nEarly stopping at epoch {epoch+1}")
        #   break

    # Load best model
    model.load_state_dict(best_model)

    return model, history, best_epoch
```

```python
# @title è¨“ç·´éç¨‹ & æœ€çµ‚è©•ä¼°å‡½å¼

def plot_training_curves(history, best_epoch):
    """
    ç¹ªè£½è¨“ç·´éç¨‹ä¸­çš„ Loss å’Œ Accuracy æ›²ç·š
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

    epochs = range(1, len(history['train_loss']) + 1)

    # --- åœ– 1: Loss æ›²ç·š ---
    ax1.plot(epochs, history['train_loss'], 'b-', label='Training Loss')
    ax1.plot(epochs, history['val_loss'], 'r-', label='Validation Loss')
    ax1.set_title('Training & Validation Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    # æ¨™è¨˜æœ€ä½³ Epoch
    ax1.axvline(best_epoch + 1, color='g', linestyle='--', label=f'Best Model (Epoch {best_epoch+1})')
    ax1.legend()
    ax1.grid(True)

    # --- åœ– 2: Accuracy æ›²ç·š ---
    ax2.plot(epochs, history['train_acc'], 'b-', label='Training Accuracy')
    ax2.plot(epochs, history['val_acc'], 'r-', label='Validation Accuracy')
    ax2.set_title('Training & Validation Accuracy')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    # æ¨™è¨˜æœ€ä½³ Epoch
    ax2.axvline(best_epoch + 1, color='g', linestyle='--', label=f'Best Model (Epoch {best_epoch+1})')
    ax2.legend()
    ax2.grid(True)

    plt.tight_layout()
    plt.show()

def generate_evaluation_report(model, test_dataset, batch_size=64, class_names=['Not Top-9', 'Top-9']):
    """
    å°æœ€ä½³æ¨¡å‹ç”¢ç”Ÿæœ€çµ‚è©•ä¼°å ±å‘Š (åˆ†é¡å ±å‘Šã€æ··æ·†çŸ©é™£ã€ROC æ›²ç·š)
    """
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)
    model.eval()

    test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size)

    all_labels = []
    all_preds = []
    all_probs = [] # ç”¨æ–¼ ROC æ›²ç·š

    with torch.no_grad():
        for batch_data in test_loader:
            batch_x = batch_data[0].to(device)
            batch_y = batch_data[1].to(device)

            outputs = model(batch_x, task_type='classification')

            # ç²å–é æ¸¬é¡åˆ¥ (for report & matrix)
            preds = torch.argmax(outputs, dim=1)
            # ç²å– "é¡åˆ¥ 1" (Top-9) çš„æ©Ÿç‡ (for ROC)
            probs = outputs[:, 1]

            all_labels.extend(batch_y.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())

    # --- 1. ç”¢ç”Ÿ "åˆ†é¡å ±å‘Š" (Precision, Recall, F1) ---
    print("="*60)
    print("Classification Report")
    print("="*60)
    report = classification_report(all_labels, all_preds, target_names=class_names)
    print(report)

    # --- 2. ç”¢ç”Ÿ "æ··æ·†çŸ©é™£" ---
    print("\n" + "="*60)
    print("Confusion Matrix")
    print("="*60)
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.ylabel('Actual Label')
    plt.xlabel('Predicted Label')
    plt.show()

    # --- 3. ç”¢ç”Ÿ "ROC æ›²ç·š" ---
    print("\n" + "="*60)
    print("ROC Curve")
    print("="*60)
    fpr, tpr, thresholds = roc_curve(all_labels, all_probs)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:0.3f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.show()
```

# è¼‰å…¥è³‡æ–™

```python
# è·¯å¾‘è¨­å®š
PROJECT_PATH = "/content/drive/MyDrive/Colab Notebooks/é›»è…¦è¦–è¦ºä¹‹æ·±åº¦å­¸ç¿’/final_project"

# è¼¸å…¥è·¯å¾‘ (Data_OHLC):
_subfolder_o = os.path.join(PROJECT_PATH, "Data_OHLC")

# è¼¸å‡ºè·¯å¾‘ (æ¨¡å‹å„²å­˜):
base_output_dir = os.path.join(PROJECT_PATH, "CNN_model")
os.makedirs(base_output_dir, exist_ok=True)

# --- è¼‰å…¥è³‡æ–™ ---
print(f"å¾ {_subfolder_o} è¼‰å…¥è³‡æ–™...")
# CNN X
X_train = np.load(os.path.join(_subfolder_o, 'X_train_CNN.npy'))
X_val = np.load(os.path.join(_subfolder_o, 'X_val_CNN.npy'))
# CNN y
y_train = pd.read_csv(os.path.join(_subfolder_o, 'y_train_CNN.csv'))
y_val = pd.read_csv(os.path.join(_subfolder_o, 'y_val_CNN.csv'))

print(f"X_train shape: {X_train.shape}")
print(f"X_val shape: {X_val.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_val shape: {y_val.shape}")
```

# æ¨¡å‹è¨“ç·´

```python
# @title Train CNN Model

# --- å»ºç«‹ TensorDataset ---
train_dataset = make_tensor_dataset(X_train, y_train, mode="threshold", threshold=0.003)
val_dataset = make_tensor_dataset(X_val, y_val, mode="threshold", threshold=0.003)

# --- è¨“ç·´è¨­å®š ---
# num_epochs = 50
learning_rates = [1e-4]
num_epochs = 30
batch_size = 64
patience = 2
weight_decay = 1e-4

# --- Grid Search for Best Learning Rate ---
best_lr = None
best_val_loss = float('inf')
best_model_state = None
best_history = None
best_epoch_num = 0

for lr in learning_rates:
    print(f"\n æ¸¬è©¦ Learning Rate: {lr}")
    model = CNN20d(num_classes=2)

    model, history, best_epoch = train_model(
        model,
        train_dataset,
        val_dataset,
        mode='classification',
        num_epochs=num_epochs,
        patience=patience,
        learning_rate=lr,
        batch_size=batch_size,
        weight_decay=weight_decay
    )

    # é¡¯ç¤ºè©² LR çš„è¨“ç·´æ›²ç·š
    print(f"\n--- è¨“ç·´æ›²ç·š (LR = {lr}) ---")
    plot_training_curves(history, best_epoch)

    current_val_loss = history['val_loss'][best_epoch]
    print(f"å®Œæˆ LR={lr} çš„è¨“ç·´ï¼Œæœ€ä½³ Val Loss={current_val_loss:.6f} (åœ¨ Epoch {best_epoch+1})")

    if current_val_loss < best_val_loss:
        best_val_loss = current_val_loss
        best_lr = lr
        best_model_state = deepcopy(model.state_dict())
        best_history = deepcopy(history)
        best_epoch_num = best_epoch

print(f"\n Grid Search å®Œæˆï¼šæœ€ä½³ LR={best_lr}, æœ€ä½³ Val Loss={best_val_loss:.6f}")
```

```python
# @title æœ€çµ‚æ¨¡å‹è©•ä¼°

# --- 1. è¼‰å…¥ Grid Search çš„æœ€ä½³æ¨¡å‹ ---
print(f"\nğŸš€ è¼‰å…¥æœ€ä½³æ¨¡å‹ (LR={best_lr}, Epoch={best_epoch_num+1})")
final_model = CNN20d(num_classes=2)
final_model.load_state_dict(best_model_state)
```

```python
# --- 2. é¡¯ç¤º "æœ€ä½³æ¨¡å‹" çš„è¨“ç·´æ›²ç·š ---
print(f"\n--- æœ€ä½³æ¨¡å‹çš„è¨“ç·´æ›²ç·š (LR = {best_lr}) ---")
plot_training_curves(best_history, best_epoch_num)
```

```python
# --- 3. ç”¢ç”Ÿ "æœ€ä½³æ¨¡å‹" çš„å®Œæ•´è©•ä¼°å ±å‘Š ---
print(f"\n--- æœ€ä½³æ¨¡å‹çš„æœ€çµ‚è©•ä¼°å ±å‘Š (on Validation Set) ---")
# æˆ‘å€‘ä½¿ç”¨ val_dataset é€²è¡Œæœ€çµ‚è©•ä¼°
generate_evaluation_report(final_model, val_dataset, batch_size=batch_size)
```
